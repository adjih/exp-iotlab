
The "radio-map" experiment(s): the data
----------------------------

The basic burst is as follows:
  * every node listen
  * one node sends a burst of packets in succession, with, for each of them:
    * "energy detection" (sample https://github.com/hikob/openlab/blob/master/appli/tests/phy/ed.c)
    * send one packet with content (no IEEE 802.15.4 header):
      * magic (16-bits field with a constant)
      * crc-32 
      * burst identifier 
      * sequence number (packet index: 0, 1, 2, 3, ...)
    * wait for fixed delay
    * [next packet]
  * every other node records received packets (and errors) in a array
    (sequence number, lqi, rssi, start_time, stop_time) XXX


The basic experiment is essentially several bursts, as follows (actually split in sub-experiences):
  * for every power:
  * for every channel:
  * for every node U:
    * clear all receiver information/array
    * make U send a burst
    * collect all receiver information 
    * store all data in a file

Burst parameters:
  * power (default: 0 dBm)
  * channel 
  * packet size (default: 50 bytes)
  * delay (default: 1 ms)
  
Information from every burst:
  * In current version, results of bursts are stored in arrays indexed as (power-index, channel-index, sender-index) where for "power-index = 2" means thats the power used is the third one in "powerList", i.e. "powerList[2]". The result is itself a multidimensionnal array with:
    * "ed": array of the "energy detection" level at the sender, before every packet
    * statistics for every node (as a receiver), hence in an sub-array indexed by receiver-index:
      * for every packet (hence in sub-array indexed by sequence-number)
        * "recv": boolean (0/1), whether the packet was received
        * "rssi": what was the rssi (if the packet was received)
        * "lqi": what was the lqi (if the packet was received)
      * also error counts per receiver, hence in a sub-array indexed by receiver-index:
        * "radio-crc": crc error, sent by radio driver from openlab
        * "radio-length": error in packet length, send by radio driver from openlab
        * "magic": packet received without the right constant field
        * "user-crc32": packet with crc32 field not matching the crc32 of the packet
        plus counts of errors detected from parsing:
        * "bad-seq-num": the sequence-number is inconsistent (too big, or not in order)
        * "invalid-handle-irq": message sometimes displayed
        * "outside-burst": packet received outside of a burst (there is a command "lock" after a burst, to instruct stop keeping track of packets)
        * "generic": misc error count reported when collecting data (e.g. packet buffer full etc.)
        * "magic2": same as "magic", only directly counted at receiver

The after all the parsing steps the results are available in the experiment directory as 
compressed "numpy" files with compressed arrays(http://docs.scipy.org/doc/numpy/reference/generated/numpy.savez_compressed.html):
 * "stat.npz" contains:
   * "ed" (int16) array indexed by (power-index, channel-index, sender-index,  sequence-number)
   * "recv" (uint8) array indexed by (power-index, channel-index, sender-index, receiver-index, sequence-number)
   * "rssi", "lqi": same indices as "recv"
 * "error.npz" contains:
   * all the error counts with the names indicated previously with index (power-index, channel-index, sender-index, receiver-index)
 * "meta.pydat", "ressources.pydat" hold all information relative to the experiment (channel list, power list, node/address list, etc.)


### Per channel information (same info is in python parseRadioExp.py summary [dirname])

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from pylab import *
```

```python
# Load data files
#expDir = "exp-2014-08-21-03h06m42"
expDir =  "exp-2015-02-04-23h02m50" # grenoble - 2015
#expDir = "exp-2015-02-04-19h19m00" # lille - 2015
expStat = np.load(expDir+"/stat.npz")
expError = np.load(expDir+"/error.npz")
```

```python
# Read meta-data, e.g. information about the "experiment"
def readPydat(name):
    with open(name, "rb") as f:
        return eval(f.read())
    
meta = readPydat(expDir+"/meta.pydat")
channelList = meta["channelList"]
powerList = meta["powerList"]
```

(the following hack was for checking loss rate anomalies [around 80%], indeed they disappear on quiet channels --- TODO: on channel 25, with it, the matching RSSI-vs-LQI is incredible, check if correct)

```python

if False:
    # HACK: copying stat data from one channel to all channels
    # so that all channels appear to have the same data.
    # (a way to: do statistics on only one channel, without changing one line of code below)
    print "\nchannel data replication hack enabled!!!\n"
    oldExpStat = expStat
    expStat = {}
    for name in ["recv","lqi","rssi","ed"]:
        expStat[name] = oldExpStat[name].copy()

    refChIdx = 25-11 ; desc="quiet-channel" # quiet channel
    #refChIdx = 22-11 ; desc="noisy-channel" # noisy channel
    for powerIdx in range(len(powerList)):
        for channelIdx in range(len(channelList)):
            if channelIdx != refChIdx:
                for name in ["recv","lqi","rssi","ed"]:
                    expStat[name][powerIdx][channelIdx] = expStat[name][powerIdx][refChIdx]
                    sys.stdout.write(".")
else: 
    desc="all-channels"
    print "Multichannel"
```

```python
# Copy data (because np.load re-loads compressed file on demands IIRC)
# Indices of the matrices are as follows:
#   (power, channel, sender, receiver, packetIdx)
#       0      1        2       3         4     
recv = expStat["recv"]#.copy()

# Hack (2015) for one set of data with transmitter error
edStat = expStat["ed"]
edStat = np.maximum(edStat, -91)
```

In the following, we plot the proportion of receivers (e.g. the percentage of nodes of the network that receive a transmission), averaged on all transmissions done during the experiment, but on a per-channel. In other terms, a value of 0.1 means that 10% of the nodes on average receive a transmission.

```python
successRate = recv[0].mean(3)
channelSuccessRate = successRate.mean((1,2))
plt.ylim(0, max(channelSuccessRate))
plt.bar(channelList, channelSuccessRate, align="center")
plt.title("average prop. receivers for one transmission [1=all nodes], per channel")
plt.xlabel("channel")
```

```python
minEnergy = edStat.min()
print ("minimum energy:", minEnergy)
energy = edStat[0].mean(axis=(1,2))
energy = energy - minEnergy # so as to get positive values 
#plt.ylim(energy.min(), energy.max())
plt.bar(channelList, energy, align="center")
plt.title("average energy sensed before sending, per channel \n(the value is the difference with minimum ever observed)")
plt.grid()
plt.xlabel("channel")
```

## Link packet loss statistics

(the statistics are taken by considering all links, e.g. pairs of (sender,receiver) on all channels)

```python
# Number of packets sent
nbPacket = recv.sum(4).max()
assert nbPacket == meta["nbPacket"]
nbPacket
```

```python
assert len(powerList) == 1
linkRecv = recv[0].sum(3, dtype=np.uint32) # linkRecv[(channel, sender, receiver)] -> number of received pkt
flatLinkRecv = linkRecv.flatten()
flatLinkRssiAvg = np.ma.masked_array(expStat["rssi"], mask=(recv == 0))[0].mean(3).flatten()

linkCountAll = (flatLinkRecv == nbPacket).sum()
linkCountNone = (flatLinkRecv == 0).sum()
linkCountSome = ((flatLinkRecv < nbPacket) & (flatLinkRecv > 0)).sum()
assert linkCountSome+linkCountAll+linkCountNone == len(flatLinkRecv)

print ("nb links with 100%% received (e.g. not even one packet missed): % 7d" % linkCountAll )
print ("nb links with   0%% received (e.g. all packets missed):         % 7d" % linkCountNone )
print ("other links:                                                   % 7d" % linkCountSome )
print ("")
print ("proportion of lossy links: %s %%" % (linkCountSome /float (linkCountAll+linkCountSome)* 100)) 
```

```python
# Direct histogram in log scale
h1 = plt.hist(flatLinkRecv, normed=True, log=True, bins=99)
```

```python
# KDE density histogram
linkLoss = stats.gaussian_kde(flatLinkRecv)
xList = np.linspace(0,nbPacket,100)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_yscale("log")
ax.plot(xList, linkLoss(xList))
ax.set_ylabel("count (logscale)")
```

```python
lossyLinkIdx = np.nonzero(((flatLinkRecv < nbPacket) & (flatLinkRecv > 0)))[0]
#lossyLinkIdx = np.nonzero(((flatLinkRecv < nbPacket) & (flatLinkRecv > 0) & (flatLinkRssiAvg < -85)) )[0]
lossyLinkRecv = flatLinkRecv[lossyLinkIdx]
h2 = plt.hist(lossyLinkRecv, normed=False, bins=99)
plt.title("histogram of the 'count of recv. pkts' \non all links [= pairs (xmit,recv)]\nFOR NON-ZERO/NON-PERFECT RECV. COUNT")
plt.xlabel("number of received packets")
plt.ylabel("number of links")
```
